```python
from lingua import Language, LanguageDetectorBuilder

detector = LanguageDetectorBuilder.from_all_languages().with_minimum_relative_distance(0).build()

texts = [
    "Tjena! What's your name? Jag är nyfiken",
    "Holla! What's your name?",
    "Hi, can I buy some äpplen?",
    "Hi, what's yoru name?",
]

for text in texts:
    print(f"Full text: {text}")
    results = detector.detect_multiple_languages_of(text)
    confidence_values = detector.compute_language_confidence_values(text)
    
    is_all_english = all(result.language == Language.ENGLISH for result in results)

    print(confidence_values)
    if is_all_english:
      print("All English!")
    else:
      print("Mixed or non-English detected!")
      for result in results:
          snippet = text[result.start_index:result.end_index]
          print(f"{result.language.name}: '{snippet}'")
    print("\n")
```

```python
from lingua import Language, LanguageDetectorBuilder

def detect_if_pure_english(text):
  detector = LanguageDetectorBuilder.from_all_languages().build()
  results = detector.detect_multiple_languages_of(text)

  # Check if all detected segments are English
  if all(result.language == Language.ENGLISH for result in results):
      return Language.ENGLISH
  else:
      return None

# Test it
text1 = "Tjena! What's your name? Jag är nyfiken"
text2 = "Holla! What's your name?"

print(detect_if_pure_english(text1))  # None
print(detect_if_pure_english(text2))  # Language.ENGLISH
```
